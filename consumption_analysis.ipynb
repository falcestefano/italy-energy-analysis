```python
import pandas as pd
from pathlib import Path 
from sqlalchemy import create_engine 
from dotenv import load_dotenv 
import os 
```

```python
raw_data = Path(r'C:\Users\falce\OneDrive\Desktop\Progetti_Data_analyst\Analisi_energia\Consumo')
```

```python
files = list(raw_data.glob('*.xlsx'))
```

```python
all_files = []

for file in files:
    df = pd.read_excel(file)
    r_file = df.dropna()
    all_files.append(r_file)

print(all_files)
```

```python
df_unito = pd.concat(all_files, ignore_index=True)
df_unito['Anno'] = df_unito['Anno'].astype(int) 
df_unito.info()
```

```python
load_dotenv(raw_data.parent/'db.env') 

user = os.getenv('DB_USER')
password = os.getenv('DB_PASSWORD')
host = os.getenv('DB_HOST')
port = os.getenv('DB_PORT')
database = os.getenv('DB_NAME')

engine_url = f'postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}'
```

```python
engine = create_engine(engine_url)
```

```python
df_unito.columns = df_unito.columns.str.lower()
df_unito = df_unito.rename(columns={'anno': 'year',
                                'regione': 'region',
                                'provincia': 'province',
                                'settore': 'sector',
                                'consumo (gwh)': 'consumption_gwh'})
```

```python
df_unito['region'] = df_unito['region'].str.lower()
```

```python
df_unito.to_csv(raw_data.parent/'energy_consumption.csv', index=False)
```

```python
df_unito.to_sql('consumption_data', con=engine, if_exists='append', index=False)
```
